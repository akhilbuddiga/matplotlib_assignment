```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/akhil/Downloads/virat kohli.xlsx')
```

```python
df
```

```python
fig, axes = plt.subplots(1,2, figsize=(30,10), dpi=(250))
axes[0].bar(df['YEAR'], df['RUNS'], color='black')
axes[0].set_xlabel('YEARS')
axes[0].set_ylabel('RUNS')
axes[0].set_title('Runs Scored Over Years');

axes[1].bar(df['YEAR'], df['AVG'], color='pink')
axes[1].set_xlabel('YEARS')
axes[1].set_ylabel('AVERAGE')
axes[1].set_title('Batting Average Over Years');
```
